```python
import numpy as np
import cv2

img = np.zeros((500, 500, 3), dtype=np.uint8)

bounding_box_list = [[[0, 100], [100, 200]],
                     [[50, 150], [150, 250]],
                     [[75, 175], [175, 275]],
                     [[100, 100], [200, 200]],
                     [[150, 120], [250, 220]],
                     [[200, 150], [300, 250]],
                     [[240, 170], [340, 270]],
                     [[240, 200], [340, 300]],
                     [[200, 250], [300, 350]],
                     [[250, 250], [350, 350]],
                     [[260, 300], [360, 400]],
                     [[280, 350], [380, 450]],
                     [[300, 400], [400, 500]]]

cv2.line(img, (0, 160), (500, 160), (100, 100, 100), 5)
cv2.line(img, (0, 320), (500, 320), (100, 100, 100), 5)

for i, bounding_box in enumerate(bounding_box_list):
    cv2.rectangle(img, (bounding_box[0][0], bounding_box[0][1]), (bounding_box[1][0], bounding_box[1][1]),
                  (255, 0, 0), 5)

for i, bounding_box in enumerate(bounding_box_list):
    point_x = int((bounding_box[0][0] + bounding_box[1][0])/2)
    point_y = max(bounding_box[0][1], bounding_box[1][1])
    
    try:
        cv2.line(img, (point_x, point_y), (old_x, old_y), (255, 255, 255), 3)
    except:
        pass
    
    old_x = point_x
    old_y = point_y

for i, bounding_box in enumerate(bounding_box_list):
    if i == 0:
        cv2.circle(img, (int((bounding_box[0][0] + bounding_box[1][0])/2), max(bounding_box[0][1], bounding_box[1][1])), 7, (0, 255, 255), 3)
    if i == (len(bounding_box_list)-1):
        cv2.circle(img, (int((bounding_box[0][0] + bounding_box[1][0])/2), max(bounding_box[0][1], bounding_box[1][1])), 7, (255, 255, 0), 3)
    else:
        cv2.circle(img, (int((bounding_box[0][0] + bounding_box[1][0])/2), max(bounding_box[0][1], bounding_box[1][1])), 3, (0, 0, 255), 3)

cv2.imshow('img', img)
cv2.waitKey()
```

```python
# (1) 탐지 되지 않는 쇼잉 화면 보여주기 코드 만들기 -> 율이형님 요청
# (2) 파란 눈금선 좌측으로 이동

import numpy as np
import cv2

capture = cv2.VideoCapture(0)

w = capture.get(cv2.CAP_PROP_FRAME_WIDTH)
h = capture.get(cv2.CAP_PROP_FRAME_HEIGHT)

danger_range_ratio = float(input("danger_range_ratio 입력 : [0.0 - 1.0]"))
danger_range_ratio = 1.0 - danger_range_ratio

while True:
    ret, frame = capture.read()    
   
    danger_range_img = np.zeros((frame.shape[0], frame.shape[1], 3), dtype=np.uint8)
    cv2.rectangle(danger_range_img, (0,int(h*danger_range_ratio)), (int(w), int(h)), (0, 0, 255), -1)
    
    cv2.line(frame, (int(w/2), 0), (int(w/2), int(h)), (255, 255, 255), 3)
    cv2.line(frame, (0, int(h/2)), (int(w), int(h/2)), (255, 255, 255), 3)
    
    for i in range(11):
        recog = int(frame.shape[0]/10)
        cv2.line(frame, (0, recog*i), (int(w/20), recog*i), (255, 0, 0), 3)
        cv2.line(frame, (int(w), recog*i), (int(w) - int(w/20), recog*i), (255, 0, 0), 3)
    
    frame = cv2.addWeighted(frame, 1.0, danger_range_img, 0.2, 0.0)
    
    cv2.imshow('cam', frame)
    if cv2.waitKey(30) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
# 낮 -> 밤 바꾸기

import cv2
import numpy as np
import matplotlib.pyplot as plt


# 이미지 로드
path = '/Users/jongya/Desktop/test_noon.jpg'
img = cv2.imread('/Users/jongya/Desktop/test_noon.jpg')
img_cvt = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)

# 이미지에 노이즈 추가
noise = np.zeros_like(img_cvt)
cv2.randn(noise, 20, 50)
img_cvt += noise

# 밝기 조절 : (1) 먼 곳은 더 어둡도록 (2) 화면 주변부는 더 어둡도록
for i in range(img_cvt.shape[0]):
    beta = (100 / img_cvt.shape[0]) * (img_cvt.shape[0] - i)
    alpha = min(i * 2 / img_cvt.shape[0], 1)
    for j in range(img_cvt.shape[1]):
        gamma = ( j - (img_cvt.shape[1]/2) ) / img_cvt.shape[1]
        img_cvt[i, j] = max((alpha * img_cvt[i, j]) * (1 - gamma**2) - beta, 0)

# img 밝기 최대값 제한
img_cvt = np.clip(img_cvt, 0, 200).astype(np.uint8)
# img ndarray int 형태로 변환
img_cvt = img_cvt.astype(np.uint8)

# 이미지 보기
plt.figure(figsize = (15, 10))
plt.subplot(211)
plt.imshow(cv2.cvtColor(img_cvt, cv2.COLOR_BGR2RGB))
plt.subplot(212)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# 이미지 저장하려면 아래 코드 실행
save_path = '저장할 경로'
cv2.imwrite(save_path, img_cvt)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/Users/jongya/Desktop/test_noon.jpg')

# Convert the image to grayscale
img_cvt = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)

# Add some noise to the image
noise = np.zeros_like(img_cvt)
cv2.randn(noise, 0, 10)
img_cvt += noise

# Apply a night filter
for i in range(img_cvt.shape[0]):
    beta = (100 / img_cvt.shape[0]) * (img_cvt.shape[0] - i)
    alpha = min(i * 2 / img_cvt.shape[0], 1)
    for j in range(img_cvt.shape[1]):
        gamma = ( j - (img_cvt.shape[1]/2) ) / img_cvt.shape[1]
        img_cvt[i, j] = max((alpha * img_cvt[i, j]) * (1 - gamma**2) - beta, 0)

# Convert the image back to color and adjust the white balance
img_cvt = np.clip(img_cvt, 0, 255).astype(np.uint8)
img_cvt = cv2.cvtColor(img_cvt, cv2.COLOR_GRAY2BGR)
lab = cv2.cvtColor(img_cvt, cv2.COLOR_BGR2LAB)
lab[:,:,1] = np.clip(lab[:,:,1] - 10, 0, 255)
lab[:,:,2] = np.clip(lab[:,:,2] - 20, 0, 255)
img_cvt = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

# Display the original and processed images
plt.figure(figsize=(15, 10))
plt.subplot(211)
plt.imshow(cv2.cvtColor(img_cvt, cv2.COLOR_BGR2RGB))
plt.title('Nighttime Image')
plt.subplot(212)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Daytime Image')
plt.show()
```

```python
# 이미지 캡쳐, 비디오 저장 기능 구현한 코드

import cv2
import numpy as np

capture = cv2.VideoCapture(0)

width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

capture.set(cv2.CAP_PROP_FRAME_WIDTH, width)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
fps = capture.get(cv2.CAP_PROP_FPS)

basic_path = '/Users/jongya/Desktop/capture/'
file_name = 0

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(basic_path + 'vid.mp4', fourcc, int(fps), (width, height), True)

print('ok')

while True:
    ret, frame = capture.read()
    if ret:
        cv2.line(frame, (0, int(height/2)), (width, int(height/2)), (255, 255, 255), 2)
        cv2.line(frame, (int(width/2), 0), (int(width/2), height), (255, 255, 255), 2)
        cv2.imshow('vid', frame)
        out.write(frame)
        if cv2.waitKey(int(1000/fps)) == ord('c'):
            cv2.imwrite(basic_path + str(file_name) + '.jpg', frame)
            file_name += 1
        if cv2.waitKey(int(1000/fps)) == ('q'):
            break

out.release()
capture.release()
cv2.destroyAllWindows()
```

```python
# 위치 따기

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

path = '/Users/jongya/Desktop/capture/'

img_list = os.listdir(path)
img_list = [ path + x for x in img_list if x.endswith('.jpg') ]
print(img_list)

index = 4
img = cv2.imread(img_list[index])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

temp_list_1 = []
temp_list_2 = []
temp_list_3 = []
temp_list_4 = []

for i in range(img.shape[1]):
    temp_y = 1000
    brightness = 500
    b = int(img[temp_y, i, 0])
    g = int(img[temp_y, i, 1])
    r = int(img[temp_y, i, 2])
    if ( b + g + r ) >= brightness:
        temp_list_1.append(i)
        
horizon_start = min(temp_list_1)        
     
for i in range(horizon_start, img.shape[1]):
    temp_y = 1000
    brightness = 500
    b = int(img[temp_y, i, 0])
    g = int(img[temp_y, i, 1])
    r = int(img[temp_y, i, 2])
    if ( b + g + r) <= brightness:
        temp_list_2.append(i)

horizon_end = min(temp_list_2)

for i in range(img.shape[0]):
    temp_x = 1036
    brightness = 600
    b = int(img[img.shape[0] - 1 - i, temp_x, 0])
    g = int(img[img.shape[0] - 1 - i, temp_x, 1])
    r = int(img[img.shape[0] - 1 - i, temp_x, 2])
    if (b + g + r) >= brightness:
        temp_list_3.append(img.shape[0] - 1 - i)

vertical_end = max(temp_list_3)

for i in range(vertical_end):
    temp_x = 1020
    brightness = 530
    gamma = 1
    b = int(img[vertical_end - i - gamma, temp_x, 0])
    g = int(img[vertical_end - i - gamma, temp_x, 1])
    r = int(img[vertical_end - i - gamma, temp_x, 2])
    if (b + g + r) <= brightness:
        temp_list_4.append(vertical_end - i)

vertical_start = max(temp_list_4)
        
print('horizon_start : ', horizon_start)
print('horizon_end : ', horizon_end)
print('vertical_start : ', vertical_start)
print('vertical_end : ', vertical_end)

cv2.line(img, (horizon_start, vertical_end), (horizon_end, vertical_end), (255, 0, 0), 5)
cv2.line(img, (horizon_start, vertical_start), (horizon_start, vertical_end), (255, 0, 0), 4)
cv2.rectangle(img, (0, 0), (600, 300), (255, 255, 255), -1)
cv2.putText(img, f'width : {img.shape[1]}, height : {img.shape[0]}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX| cv2.FONT_ITALIC, 1, (0, 0, 0), 2)
cv2.putText(img, f'horizon_start : {horizon_start}', (10, 80), cv2.FONT_HERSHEY_SIMPLEX| cv2.FONT_ITALIC, 1, (0, 0, 0), 2)
cv2.putText(img, f'horizon_end : {horizon_end}', (10, 130), cv2.FONT_HERSHEY_SIMPLEX | cv2.FONT_ITALIC, 1, (0, 0, 0), 2)
cv2.putText(img, f'vertical_start : {vertical_start}', (10, 180), cv2.FONT_HERSHEY_SIMPLEX | cv2.FONT_ITALIC, 1, (0, 0, 0), 2)
cv2.putText(img, f'vertical_end : {vertical_end}', (10, 230), cv2.FONT_HERSHEY_SIMPLEX | cv2.FONT_ITALIC, 1, (0, 0, 0), 2)
cv2.putText(img, f'height - object_bot_center : {img.shape[0] - vertical_end}', (10, 280), cv2.FONT_HERSHEY_SIMPLEX | cv2.FONT_ITALIC, 1, (0, 0, 0), 2)

plt.figure(figsize = (20, 10))
plt.title(img_list[index].replace(path, ''), fontdict={'size':20})
plt.imshow(img)
plt.show()
```
