# 필요 라이브러리 import

```python
import pandas as pd
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# 데이터셋 가져와서 cancer 변수에 넣기
cancer = load_breast_cancer(as_frame = True)
```

```python
# 데이터 살펴보기
print(" === cancer === ")
print(cancer)

print(" === cancer.data === ")
print(cancer.data.head(10))

print(" === cancer target === ")
print(cancer.target.value_counts())
# 0 : 암이 아님
# 1 : 암

print(" === cancer target name === ")
print(cancer.target_names)
```

```python
# 모델들 생성  
## 모델은 (1) 로지스틱 회귀 (2) K 근접 분류기 두 가지를 이용하며
## 모델들을 Voting 분류기를 통해 가장 좋은 값을 낼 예정이다.
lr_clf = LogisticRegression(solver='liblinear') # 로지스틱 회귀 모델
knn_clf = KNeighborsClassifier(n_neighbors=8) # K 근접 분류기 (초기 K값(구심값) = 8로 설정)

## 보팅 모델 생성
vo_clf = VotingClassifier( estimators=[('lr',lr_clf),('knn',knn_clf)], voting='soft')
### n 개의 모델을 넣어줄 수 있으며, 파라미터는 [(1번분류기표기명, 1번분류기), ..., [n번분류기표기명, n번분류기)] 형식이 된다.
### Voting 방식은 soft 로 설정
```

```python
# 데이터셋 분리

X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                    cancer.target,
                                                    test_size=0.2,
                                                    random_state=156)
```

```python
# voting 모델
## 모델들에 학습시키기
vo_clf.fit(X_train, y_train)

## 학습 된 모델에 대한 예측값 내기
pred = vo_clf.predict(X_test)

## 정확도 확인하기
print("모델명 : voting 정확도 : ", accuracy_score(y_test, pred))
```

```python
# lr, knn 모델 각각
models = [lr_clf, knn_clf]
for model in models:
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print( f"모델명 : {model.__class__.__name__} 정확도 : {accuracy_score(y_test, pred)}")
```
