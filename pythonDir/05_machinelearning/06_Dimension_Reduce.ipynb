# 차원 축소  
  
* 차원(변수=feature) 축소를 한다는 것.  
* 목적 : 다중공선성 감소, 데이터 해석력 증가  
* 단계 : (1) 피처 선택 (2) 피처 추출 

# 주성분 분석 PCA (Principal Component Analysis)  
  
* 차원 축소가 진행되면, 모델의 예측도는 떨어지는 경우가 많다.  
* 변수가 줄어듦으로 어쩔 수 없으나,  
* 반대급부로 다중공선성 방지와 설명력 증가가 장점이다.  

## 데이터 준비

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris(as_frame=True)
df = iris.data
```

```python
df['target'] = iris.target
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
```

```python
df.head(2)
```

## 데이터 분포 확인

```python
# sepal 분포 확인

markers=['^', 's', 'o']

for i, marker in enumerate(markers):
    x = df[df['target'] == i ]['sepal_length']
    y = df[df['target'] == i ]['sepal_width']
    plt.scatter(x, y, marker=marker, label=iris.target_names[i])

plt.legend()
plt.show()
```

```python
# petal 분포 확인

markers=['^', 's', 'o']

for i, marker in enumerate(markers):
    x = df[df['target'] == i ]['petal_length']
    y = df[df['target'] == i ]['petal_width']
    plt.scatter(x, y, marker=marker, label=iris.target_names[i])

plt.legend()
plt.show()
```

## 데이터 전처리 (스케일링)

```python
## Standard Scaler 처리
from sklearn.preprocessing import StandardScaler

df_scaled = StandardScaler().fit_transform(df.iloc[:,:-1])
# 지금은 feature가 4개
```

## 차원 축소

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # 원래 n개의 칼럼인 데이터를 2개의 칼럼으로 차원축소하겠다.
iris_pca = pca.fit_transform(df_scaled)  #.fit 과 .transform을 함께 하는 syntax
```

```python
iris_pca # 4개였던 feature가 2개로 줄어들었음을 알 수 있다.
```

```python
# pca 데이터프레임 만들기
pca_columns = ['pca_com_1', 'pca_com_2']
df_pca = pd.DataFrame(iris_pca, columns = pca_columns) # pca_com_1, 2라는 두 개의 칼럼명으로 
df_pca['target'] = iris.target # target은 target이라는 칼럼명으로
df_pca
```

```python
# pca_com_1, pca_com_2 분포 확인

markers=['^', 's', 'o']

for i, marker in enumerate(markers):
    x = df_pca[df_pca['target'] == i ]['pca_com_1']
    y = df_pca[df_pca['target'] == i ]['pca_com_2']
    plt.scatter(x, y, marker=marker, label=iris.target_names[i])

plt.legend()
plt.show()
```

```python
# 차원 축소 후 요인(여러 feature들이 혼합된 주성분)을 설명
pca.explained_variance_ratio_

## [0.72962445, 0.22850762] -> 데이터를 com_1이 72.9%, com_2가 22.8%이 설명한다.
## 즉, 두 주성분으로 데이터의 95% 가량을 설명할 수 있다는 말
### iris는 데이터셋이 작아 설명률이 높으나.. 현실은 그렇지 않다고 함 
```

## 교차 검증

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
# 랜덤포레스트로 원래 전체 데이터 (iris.data) 교차검증 scoring
rcf = RandomForestClassifier(random_state=156)
scores = cross_val_score(rcf, iris.data.iloc[:,:-1], iris.target, scoring='accuracy', cv=3)
print(f'개별 정확도 : {scores}, 평균 정확도 : {np.mean(scores)}')
```

```python
# 랜덤포레스트로 PCA된 데이터 (df_pca) 교차검증 scoring
rcf = RandomForestClassifier(random_state=156)
scores = cross_val_score(rcf, df_pca.iloc[:,:-1], iris.target, scoring='accuracy', cv=3)
print(f'개별 정확도 : {scores}, 평균 정확도 : {np.mean(scores)}')
```
