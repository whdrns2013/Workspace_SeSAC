인공지능  

- 인공지능 : 사람과 같은 학습, 사고가 가능한 기계를 만들기 위한  
- 머신러닝 : 기계가 스스로 학습하는  
- 딥러닝 : 사람의 뉴런과 같은 학습 방법  

머신러닝  

- 지도학습 : 답을 주고 기계를 학습시킴  
- 비지도학습 : 답을 주지 않고 하는 기계 학습  
- 강화학습 : 학습 모델간 혹은 모델 내 경쟁을 통해 학습 (오차 최소화)  

## 지도학습

답(레이블)이 함께 있는 트레이닝 데이터 셋을 통해 기계가 답을 배운 뒤  
기계가 해당 문제를 해결할 수 있는 방법을 새로운 데이터셋에 적용하는 것  

* 회귀 : 값에 대한 예측 / 연속적인 값  
* 분류 : 어떤 범주러 넣을 것이냐 / 이산 값    

### 회귀
  
* 선형 회귀 : 두 변수 간 일차방적식  
* 오차에 대한 계산 : 오차는 절대값 혹은 제곱처리를 해야 오차의 평균을 잘 알 수 있음  
* 경사하강법 : 오차 함수에서 미분을 통해 기울기가 가장 작은(혹은 0인) 지점을 찾아 함수를 구함 (미분은 기울기를 구하는 것임)  

  

### 분류  
  
* 의사결정 트리  
* K-nearist  

## 비지도학습  

트레이닝 데이터에 정답이 없고, 데이터만 있음.  
데이터 내 관측값들 간의 패턴, 특성, 유사성 등을 분석하는 목적의 학습이 주가 된다.  

### 군집  

* k-means clustering  : k 개의 점을 찍은 뒤, 가까운 관측값들을 군집화 함
