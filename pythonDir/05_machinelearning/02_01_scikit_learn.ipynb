### sklearn 살펴보기

```python
# 사이킷 런 설치
pip install scikit-learn=버전

# 사이킷 런 import
import sklearn
print(sklearn.__version__)
```

```python
# 사이킷런에 포함된 기본 데이터셋 살펴보기
# from sklearn.datasets import load_데이터셋

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier # 분류에 사용하는 DecisionTree (의사결정나무)
from sklearn.tree import DecisionTreeRegressor # 회귀에 사용하는 DecisionTree (의사결정나무)
from sklearn.model_selection import train_test_split # 훈련데이터와 테스트 데이터를 나누는 것
import pandas as pd # 판다스는 툴로써 많이 사용됨  
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
print(iris) # iris 데이터셋 살펴보기

iris_df = load_iris(as_frame=True) # 데이터 프레임 형태로 iris 데이터셋 불러옴
print(iris_df.data) # iris 데이터셋을 데이터프레임 형태로 보기

print(iris.target) # 타겟값(=label값 = 답) - numpy 형식이다.
print(iris.target_names) # 타겟값의...
```

### 훈련 데이터와 테스트 데이터 나누기 : train_test_split  
  
* train_test_split(나눌 데이터, 

```python
# 훈련 데이터와 테스트 데이터 나누기

from sklearn.model_selection import train_test_split

print("==== 데이터셋 나누기 ====")
X_train,X_test,y_train,y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=11)
# 데이터셋을 트레이닝 0.8 : 테스트 0.2로 나누고 각각의 변수에 넣는다.
print("\n")

X_train
```

```python
# 의사결정나무 분류모델을 사용 (dt_clf 변수에 할당함)
dt_clf = DecisionTreeClassifier(random_state=11)

# 모델에 트레이닝 데이터셋을 통해 훈련시킴
dt_clf.fit(X_train,y_train)
```

```python
# 의사결정나무의 predict( 예측 ) 모델을 사용한 값을 변수에 담아줌
pred = dt_clf.predict(X_test)

# 예측값과 실제 답안과 비교
list(zip(pred, y_test))
```

```python
# 위와 같이 눈으로 직접 출력해서 볼 수도 있지만, 데이터가 많으면 직접 비교하는 것은 불가
# 평가에 대한 라이브러리는 sklearn.metrics에 모두 있음 (첫 시행때에는 93.33%의 정확도를 보임)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

### 모델을 저장과 불러오기
  
* 모델 저장 -> pickle을 통해 가능

```python
# 모델 저장하기
import pickle

f = open('model', 'wb')  # 쓰기모드 : w / 바이트 형태 쓰기모드 : wb
pickle.dump(dt_clf, f)  # dt_clf 를 f 라는 변수에 dump (저장)
# pickle.dump # 파일로 뭔가를 할 때 dump
# pickle.dumps # 파일이 아닌 형태로 뭔가를 할 때 dumps

f.close() #f를 닫아줌
```

위 사항을 진행하면, 아래와 같이 model 이라는 파일 형태로 저장된다.  


```python
# 모델 불러오기

f = open('model', 'rb')
model = pickle.load(f)  # 저장했던 모델을 model 변수에 넣어준다.
model.predict([[4.67, 3.9, 2, 0.4]])  # 저장했던 모델을 그대로 사용할 수 있다.
```

## 테스트용 데이터셋 분류 및 교차 검증  
  
* 훈련용 데이터와 실제 분석 데이터가 같게 되면 "과대적합"이 발생한다.  
* 훈련용 데이터에는 잘 예측하지만, 새로운 데이터에 대해서는 정확한 예측을 하지 못하는 경우를 말한다.  
  
* 이를 방지하기 위해  
* (1) 훈련용 데이터 train data   
* (2) 테스트 데이터 test data  
* 로 나누어서 훈련과 테스트를 진행하게 되며,  
* 이로써도 부족할 수 있어 학습 데이터 셋을  
* (3) 검증 데이터 셋으로도 나눈다.  
  
* 이로써도 부족할 수 있어,  
* 훈련용 데이터를 교체한 경우와 비교하는 교차검증을 진행하기도 한다.  

```python
iris = load_iris()  # iris에 데이터셋 넣고
dt_clf = DecisionTreeClassifier()  # 사용할 모델은 의사결정나무 분류형
dt_clf.fit(iris.data, iris.target)  # iris 데이터셋을 학습하는데, iris target ...?
pred = dt_clf.predict(iris.data)  # 예측 모델에 분석할 데이터를 넣는다.
accuracy_score(iris.target,pred)  # 정확도를 보면 당연히 100% 이다. 학습 데이터와 테스트 데이터가 동일하기 때문
```
