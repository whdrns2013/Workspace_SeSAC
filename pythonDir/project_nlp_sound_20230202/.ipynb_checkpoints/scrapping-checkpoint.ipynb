### 1단계 : 메인페이지

(1) 대카테고리 : `<li class="cate_list_header">생활</li>`  
- 유형 : 텍스트  
- 연결 링크 : 없음

(2) 중카테고리 : `<li class="element"><a href="/health" title="건강">건강</a></li>`
- 유형 : element  
- 연결 링크 : /소카테고리  

### 2단계 : 게시판 페이지
https://theqoo.net/게시판명



(1) 게시글 제목 : 
```html
<td class="title">
<a href="/index.php?mid=health&amp;document_srl=2709340827"><span style="">정강이뼈 제일 위 통증</span></a>
<a href="/index.php?mid=health&amp;document_srl=2709340827#2709340827_comment" class="replyNum">1</a> </td>
```
- `<td class = "title">` 표 안에 있음  
- 첫 번째 `<a href>` 안에 게시물 링크가 있음  
- `<span style = "">` 안에 게시물 제목 텍스트가 있음  
- 두 번째 `<a href>` 안에 댓글 링크가 있음  
- `class="replyNum"` 안에 댓글 수가 있음
- 세 번째 `<td class="cate"><span style="color:">잡담</span></td>`
