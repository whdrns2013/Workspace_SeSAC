## 사용법

## Pandas 기본 함수
* `pd.read_확장자('경로')` : 해당 경로의 파일 읽기
* `pd.read_csv('경로', sep="백문자")` : 해당 경로의 csv 파일을 읽기 (구분자 = 백문자)
* `데프.head(n)` : 데이터프레임의 윗부분 n 개 자료를 추출해 보여줌
* `데프.tail(n)` : 데이터프레임의 아랫부분 n 개 자료를 추출해 보여줌
* `데프.sample(n)` : 데이터프레임에서 샘플 n 개 자료를 추출해 보여줌
* `데프.info()` : 데이터프레임의 요약 정보 반환. Null 값 여부, 칼럼 수 등
* `데프.describe()` : 데이터프레임의 수치적 값 요약 정보 반환.
* `DF.groupby('칼럼명')` : 특정 칼럼으로 그룹핑을 한다.
* `DF.replace('원값', '바꿀값')` : 데이터프레임에 있는 원값을 바꿀값으로 변환

## Pandas DF 속성
* `데프.shape` : 데이터프레임의 shape(n*m행렬) 모양새를 추출
* `데프.size` : 데이터프레임 의 값 개수 반환
* `데프.ndim` : 데이터프레임의 차원 반환
* `데프.column` : 데이터프레임의 열 제목들 반환
* `데프.index` : 데이터프레임의 인덱스 제목 (행 제목)을 반환
* `데프.values` : 데이터프레임의 value 값들을 반환
* `데프.dtypes` : 데이터프레임 안의 value 들의 데이터타입을 본다. 칼럼별로 데이터 타입은 같기에, 칼럼제목 = 데이터타입 형식으로 반환됨.

## Pandas 데이터 추출

* `DF['칼럼명']` : 특정 칼럼의 값들을 추출. (데이터타입은 Series) 칼럼은 하나만 가능. 여러 칼럼을 검색하려면 -->
* `DF[['칼럼명','칼럼명'...]]` : 특정 칼럼들의 값들을 추출. (데이터타입은 DataFrame)칼럼들을 리스트로 묶는다.
* `DF.loc['로우명', '칼럼이름']` : (칼럼,로우) 조건에 해당하는 원소값 반환
* `DF.loc['로우명')]` : 특정 로우명 행만 추출
* `DF.loc['로우인덱스', '칼럼인덱스']` : (칼럼 인덱스,로우 인덱스) 조건에 해당하는 원소값 반환
* `DF.loc['로우인덱스')]` : 특정 인덱스 행만 추출
* `DF.groupby('칼럼명')` : 특정 칼럼으로 그룹핑을 한다.

</br>

* loc, iloc 심화 사용

```
===행 X 열 에 해당하는 원소값 반환도 가능===
print(df.loc[:3, 'country'])

===행 X 열 범위로도 추출이 가능하다.===
print(df.loc[1:3, 'country':'continent'])

===행 X 열 범위로도 추출이 가능하다.===
print(df.loc[[1, 3, 5], ['country','continent','pop']])
      
===특정 행만 추출 : 인덱스 번호를 기준으로===
print(df.iloc[-1])

===특정 행만 추출 : 슬라이싱도 가능하다.===
print(df.iloc[:3])

===행 X 열 범위에 맞는 값 추출도 가능===
print(df.iloc[[0, 3, 5, 10], [1, -1]])
```


````python
# pandas 임포트와 예시 데이터 불러오기

import pandas as pd

df = pd.read_csv('data/gapminder.tsv', sep="\t")
df
````

````python
print("===특정 칼럼의 데이터 불러오기===")
print(df['country'])
print(type(df['country']))
print("\n")

print("===특정 칼럼'들'의 데이터 불러오기===")
print(df[['country', 'continent']])
print("\n")

print("===head, tail, sample 추출===")
print(df['country'].head(4))
print(df['country'].tail(4))
print(df['country'].sample(4))
print("\n")

# print("===특정 행만 추출 : 부여된 행 이름을 기준으로===")
# print(df.loc('Kenya'))
# print("\n")

print("===행 X 열 에 해당하는 원소값 반환도 가능===")
print(df.loc[:3, 'country'])
print("\n")

print("===행 X 열 범위로도 추출이 가능하다.===")
print(df.loc[1:3, 'country':'continent'])
print("\n")

print("===행 X 열 범위로도 추출이 가능하다.===")
print(df.loc[[1, 3, 5], ['country','continent','pop']])
print("\n")
      
print("===특정 행만 추출 : 인덱스 번호를 기준으로===")
print(df.iloc[-1])
print("\n")

print("===특정 행만 추출 : 슬라이싱도 가능하다.===")
print(df.iloc[:3])
print("\n")

print("===행 X 열 범위에 맞는 값 추출도 가능===")
print(df.iloc[[0, 3, 5, 10], [1, -1]])
print("\n")

print("===특정 칼럼으로 그루핑===")
print(df.groupby('country'))
print("\n")
````

## Pandas 통계

* `DF.mean(skipna:NA값 스킵할지, numeric_only:숫자값만 취급할건지 ..)` : 데이터프레임의 칼럼별 평균값 구하기
* `DF.groupby(['칼럼1','칼럼2'])['칼럼3'].mean()` : 칼럼1, 2로 그루핑한 뒤, 칼럼3 값의 평균값을 구한다.
* `DF.describe()` : 통계 요약
* `DF.median` : 중앙값

````python
# pandas 임포트와 예시 데이터 불러오기
import pandas as pd
df = pd.read_csv('data/gapminder.tsv', sep="\t")
df
````

````python
print("===데이터 프레임의 평균값 구하기===")
print(df.mean())
print("\n")

print("===평균값 구하기 : NA스킵, 숫자만===")
print(df.mean(skipna=True, numeric_only=True))
print("\n 앗 여긴 똑같네")

print("===year로 묶은 뒤 lifeExp의 평균값 구하기 : 연도별 예상 평균수명===")
print(df.groupby('year')['lifeExp'].mean())
print("\n")

print("===연도-대륙별 예상 평균수명과 GDP===")
print(df.groupby(['year','continent'])['lifeExp', 'gdpPercap'].mean())
print("\n")

print("===연도-대륙별 예상 평균수명과 GDP===")  # 이거 안되네
print(df.groupby(['year','continent']).loc['continent' == 'asia', ['lifeExp', 'gdpPercap']].mean())
print("\n")
````

````python
print("===칼럼별 uniqe 한 항목들 갯수 : nunique===")
print(df.nunique())
print("\n")

print("=== nunique + Groupby ===")
print(df.groupby('continent')['country'].nunique())
print("\n")
````

## 데이터프레임의 처리

* `random.shuffle(데이터프레임 or 특정 칼럼)` : 데이터프레임이나 특정 칼럼의 값들을 무작위로 섞어줌
* `DF.drop(indexs='인덱스', columns='칼럼', inplace='덮어쓸지 Boolean')` : 데이터프레임의 열 혹은 행 삭제 / axis로 열이냐 행이냐를 선택할 수도 있으나, 잘 사용치 않음

````python
# 랜덤 셔플
import random

print("=== 셔플 전 자료 ===")
print(df)
print('\n')

print("=== Age 칼럼 셔플 후 자료 ===")
df['Age']
random.shuffle(df['Age']) # Age 칼럼 데이터 랜덤으로 셔플
print(df)
print('\n')
````

````python
# 열 혹은 행 삭제

print("=== 열 삭제 전 자료 ===")
print(df)
print('\n')

print("=== Age 열 삭제 후 자료 ===")
df.drop(columns='Age', inplace=True)
print(df)
print('\n')
````

### 옵션

* dumps : 파일이 아닌 형태로 저장된 데이터를 저장 / 불러오기

````python
import pickle
f = open('data.pickle', 'wb')
# f 라는 변수 = data.pickle 이라는 파일을 wb 모드로 연 상태
pickle.dump(df,f)
# 피클을 통해 df라는 데이터를 f 라는 변수(=data.pickle 파일)에 dump 형태로 저장
f.close()
# 열린 f 라는 파일(=data.pickle) 닫기
````

````python
f = open('data.pickle', 'rb')
# f 라는 변수에 data.pickle 이라는 파일을 rb 모드로 열기
pickle.load(f)
# f 라는 변수 읽어오기
````

````python
import json


f = open('data.json', 'w')
json.dump(df['Name'].to_list(), f)
f.close()
f = open('data.json', 'r')
data = json.load(f)
print(data)


df_json = df.to_json()
f = open('data_2.json', 'wb')
json.dump(df_json, f)
f.close()
f = open('data_2.json', 'r')
data = json.load(f)
print(data)

````
