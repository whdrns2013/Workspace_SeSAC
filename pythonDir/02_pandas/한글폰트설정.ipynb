## 한글 폰트 사용법 (1) Font Properties 사용

* (1) 실제파일이 필요 : 폰트파일을 -> 파이썬 폴더에 넣어놓기
* (2) path 변수에 폰트 파일 주소값을 넣기
* (3) matplotlib의 font_manager 모듈을 사용하여
* (4) font properties 만들기 : `fontprop = font_manager.properties(fname='폰트파일주소', 기타 옵션들)`
* (5) 표 제목, 각 축 제목 선언시 `fontproperties=fontprop` 옵션을 사용

> font_manager 옵션
> * fname = 폰트 파일 주소값
> * size = 폰트 사이즈
  
> font properties 장단점
> * 폰트 파일이 없으면 사용 불가능
> * 파일 경로가 필요


```python
path = 'font/malgun.ttf'
```

```python
import matplotlib.font_manager as fm
fontprop = fm.FontProperties(fname=path, size=16)

data = np.random.randint(-100, 100, 60).cumsum()
plt.plot(range(60), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격', fontproperties=fontprop)
plt.xlabel('시간(분)', fontproperties=fontprop)
```

## 한글 폰트 사용법 (2) 전역글꼴 설정 rcParams로 설정

* (1) matplotlib.pyplot 라이브러리 내의 폰트 설정 메서드 사용
* (2) `matplotlib.pyplot.rcParams['font.family'] = '폰트'` : 전역 폰트 설정

> 원리
> * matplotlib 라이브러리의 폰트명 (f_name) 파일을 확인해보면
> * 그 안에 기본 폰트의 폰트명이 들어가있음을 알 수 있다.
> * 이에 더해, fontmanager 라이브러리에서 사용 가능한 폰트 종류들을 확인한다.
> * 이 종류에 포함되어있는 폰트는 사용이 가능하다.

> 장단점
> * 장점 : 파일이 필요 없음
> * 단점 : fontmanager 목록에 없는 폰트는 사용 불가능
> * 단점 : 세션이 종료되면, 원래 기본 글꼴로 복원됨
  
> 단점 극복 (목록에 없는 폰트)
> * (1) 폰트 다운로드 후 설치
> * (2) 주피터 노트북 재구동
> * (3) fontmanager 의 폰트리스트에 설치한 폰트가 추가됨을 확인
> * (4) 그래도 추가가 안되어있으면 : c:사용자:user:.matplotlib 폴더를 삭제 후 주피터 재구동
> * (5) fontmanager 의 폰트리스트에 설치한 폰트가 추가됨을 확인
> .matplotlib 폴더 : matplotlib 의 설정 내용이 저장됨. 삭제해도 다시 만들어짐. 뭔가 refresh가 안될 때에는 이 폴더를 삭제 후 재구동

```python
import matplotlib
import matplotlib.pyplot as plt

print("=== 기본 폰트 세팅 확인 ===")
print("기본 폰트 사이즈 : "), print(plt.rcParams['font.size'])
print("기본 폰트 타입 : "), print(plt.rcParams['font.family'])
print('\n')

print("=== 폰트 변경 ===")
data = np.random.randint(-100, 100, 60).cumsum()
plt.plot(range(60), data, 'r')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.plot(range(60), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')

```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib

print("=== 그럼 어떤 원리로 한글을 쓸 수 있는 걸까? ===")
print(matplotlib.matplotlib_fname()) # matplotlib의 기본 폰트 기록된 파일
fm.findSystemFonts() # font_manager에 등록되어있는 폰트 목록 보기
fm.fontManager.ttflist # font_manager에 있는 ttf 리스트

# font_list 만들기 (보기 편하게)
font_list = [(f.fname, f.name) for f in fm.fontManager.ttflist if 'Malgun' in f.name]
print(font_list)

# font_list 만들기 (보기 편하게)
font_list = [(f.fname, f.name) for f in fm.fontManager.ttflist if 'Nanum' in f.name]
print(font_list)
```
