# 데이터 전처리

* '깔끔한 데이터' 가 필요

> 깔끔한 데이터 만들기
> 1. 데이터 분석 목적에 맞는 데이터를 모아 테이블을 만들어야 한다.
> 2. 변수는 열(coluumn)로 구성하고
> 3. 측정값은 행 (row) 에 넣어야 한다.

## 데이터 연결

* pandas 의 concat 함수를 사용한다.
* `pd.concat([연결할 DF들], 기타 옵션들)` : 데이터 연결

> <옵션>  
> axis : 값이 0이면 행(row) 쪽으로 데이터를 붙이며, 값이 1이면 열(column) 쪽으로 데이터를 붙임  
>

### Concat

```python
import pandas as pd

df1 = pd.read_csv('data/concat_1.csv')
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
```

```python
# 데이터 확인
print(df1)
print(df2)
print(df3)
```

```python
# 데이터 연결
pd.concat([df1, df2, df3], axis=1)
pd.concat([df1, df2, df3], axis=0)
```

```python
new = pd.Series(['n1','n2','n3','n4'])
pd.concat([df1,new])
```

```python
new_df = pd.DataFrame([['n1','n2','n3','n4']], columns=['A', 'B', 'C', 'D'])
df1.columns
```

```python
pd.concat([df1,new_df])
```

```python
df1.append(new_df)
# concat과 동일한 기능이나, 곧 없어짐
```

```python
pd.concat([df1,new_df], axis=0, ignore_index=True)
# 행 방향으로 통합
# 원래의 index값은 무시하고 통합
```

```python
pd.concat([df1,new], join='inner')
# inner 조인
# 공통값이 없으므로, 데이터가 없다.
```

## 누락값 처리

### 누락값 확인하기

* `DF.isna` : NA값 확인
* `DF.notna` : NA값이 아닌 값
* `DF.isnull` : null 값 확인
* `DF.notnull` : null 이 아닌 값

### 누락값 처리하기
* `DF.dropna` : NA값을 DROP 하기
* `DF.fillna(값)` : NA값을 '값'으로 넣기
* `DF.interpolate()` : 누락값을 중간값으로 대치
