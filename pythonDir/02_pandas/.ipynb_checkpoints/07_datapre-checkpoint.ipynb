# 기타 참고

* 열 이름 바꾸기
* 방법1 : `자료.columns = [1번 칼럼 이름, 2번 칼럼 이름, ... n번 칼럼 이름]`
* 방법2 : `자료.rename(columns = {원래칼럼명1 : 변경칼럼명1, ... 원래칼럼명n : 변경칼럼명n})`
</br>
* inplace : 본 데이터에 덮어씌울지에 대한 결정. True = 덮어씌움
</br>
* 코드 작성시 줄바꿈 : 코드가 긴 경우 줄바꿈을 한다. 위치는 콤마 찍고 나서가 적절.

## melt : 자료를 녹이고 성형하기

* `자료 = pd.melt(데이터, 기타 메서드 인자)` : 자료를 녹이다 라는 뜻으로 보면 됨. 특정 형태의 자료를 녹여 내가 원하는 형태로 바꾸기 위한 전처리
> 메서드 인자
> * id_vars : 위치를 그대로 유지할 열(칼럼)의 이름을 지정한다.
> * value_vars : 행으로 위치를 변경할 열(칼럼)의 이름을 지정한다.
> * var_name : 하나의 칼럼으로 통합되는 칼럼의 이름을 지정한다.
> * value_name : 값으로 나오는 칼럼의 이름을 지정한다.

* 칼럼 명 바꾸기 방법
> 칼럼 명 바꾸기 방법  
> (1) `자료.columns = [1번 칼럼 이름, 2번 칼럼 이름, ... n번 칼럼 이름]`  
> (2) `자료.rename(columns = {원래칼럼명1 : 변경칼럼명1, ... 원래칼럼명n : 변경칼럼명n})`
  

### MELT 예제 1

```python
# MELT 예제 1
import pandas as pd
df = pd.read_csv('data/pew.csv')
df
```

```python
data = pd.melt(df, id_vars='religion')
data
#id_vars : 그대로 열로 유지할 칼럼
#value_vars : 행으로 변경할 칼럼. 지정하지 않으면 모두 행으로 변경된다.
```

```python
print(" === 칼럼 이름 변경 전 === ")
print(data.columns)
print("\n")

print(" === 칼럼 이름 변경 후 === ")
data.columns = ['religion', 'income', 'count']
print(data)
print("\n")

print(" === 칼럼 이름 변경 -다른 방법 === ")
data = pd.melt(df, id_vars='religion', var_name='income', value_name='count')
print(data)
print("\n")

print(" === 칼럼 이름 변경 -다른 방법2 === ")
data.rename(columns = {'income'  :'소득영역', 'count' : '결과수'}, inplace=True)
print(data)
print("\n")
```

### MELT 예제 3 : 

```python
# melt 예제 3
df = pd.read_csv('data/country_timeseries.csv')
df.sample(10)
```

```python
# melt 로 칼럼 합치기
df_melt = pd.melt(df, id_vars = ['Date', 'Day'])
df_melt
```

```python
# 특정 열의 데이터 바꿔주기
print(" === 상태와 국가명이 혼합된 variable 데이터 분리 === ")
print(df_melt.variable.str.split("_"))

print(" === 상태와 국가명을 별도의 변수에 담아주고 === ")
case = df_melt.variable.str.split("_").str.get(0)
country = df_melt.variable.str.split("_").str.get(1)

print(" === df_melt 데이터프레임에 새로운 칼럼에 담아주기 === ")
df_melt['status'] = case 
df_melt['country'] = country

df_melt
```

### Melt 예제 4 : 중복 데이터 처리하기

```python
import pandas as pd
df = pd.read_csv('data/billboard.csv')
print('column : '), print(df.columns)
print('data : '), df.sample(5)
```

```python
# df_melt 만들기
df_melt = pd.melt(df, id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        var_name='week',
        value_name='rating')

# df_melt 살펴보기
print(df_melt), print(df_melt.shape)
print(df_melt.columns)
```

```python
# Loser 라는 곡이 중복이 많다. 얼마나 될까?
df_melt[df_melt.track == 'Loser']
```

```python
# 중복을 제거하기 위해 DF를 복사 : 24092 행
df_songs = df_melt[['year', 'artist', 'track', 'time']]
print(df_songs), print(df_melt) # 첫 시도 시점에서 둘의 행 수는 같다.

# 중복값 제거
df_songs.drop_duplicates(inplace = True)
print(df_songs)
```

```python
# id값을 부여하기
df_songs['id'] = range(len(df_songs))
df_songs
```

```python
# 이후 원본 df_melt와 병합하기 (의의 = 곡별 id 값이 부여된다.)
df_ratings = df_melt.merge(df_songs, on=['year', 'artist', 'track', 'time'])
df_ratings = df_ratings[['id', 'date.entered', 'week', 'rating']]
df_ratings
```

## Pivot Table : 여러 열을 하나로 정리하기

* 피벗 테이블
> 피벗 테이블 방법  
> `DF.pivot_table()` : 피벗 테이블 만들기  
> `PV.values` : 피벗 테이블 안의 밸류들 보기
> `PV.index` : 피벗 테이블의 인덱스 보기
> `PV.reset_index()` : 피벗 테이블을 원래 데이터 형태로 돌리기 / inplace = True 시 덮어쓰기

```python
df = pd.read_csv('data/weather.csv')
df
```

```python
# melt 를 통해 날짜, 기온 데이터를 var, value 칼럼으로 만듦
df_melt = pd.melt(df, id_vars=['id', 'year', 'month', 'element'],
                  var_name='day', value_name='temp')
df_melt
```

```python
df_pivot = df_melt.pivot_table(index=['id', 'year', 'month', 'day'], columns='element', values='temp')
df_pivot
```

```python
# 피벗테이블의 밸류, 인덱스들 보기
print(df_pivot.values)
print(df_pivot.index)
```

```python
# 피벗한 것을 다시 원본으로 돌리기
df_pivot.reset_index()
# 원본에 저장하려면 inplace = True 값으로 지정하면 된다.
```
