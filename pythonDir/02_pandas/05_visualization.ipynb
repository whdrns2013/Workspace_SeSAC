### 히스토그램
----

* `DF.hist(시각화 할 데이터)` : 히스토그램
* `DF.settitle('타이틀명') ` : 제목 세팅
* `DF.setxlabel('x축명') ` : x축 세팅
* `DF.setylabel('y축명') ` : y축 세팅


* 값 유형 : 연속값 대상
* 값들을 일정 범위씩으로 잡아서 (값1 이상 ~ 값2 미만), 관측값들의 개수를 기록


```python
# 사용 대상 데이터 : seamap - tips
import seaborn as sns
tips = sns.load_dataset('tips')

fig = plt.figure()
a1 = fig.add_subplot(1,1,1)
a1.hist(tips['total_bill'])

# 3.07 ~ 7.844 : 7개
# 7.844 ~ 12.618 : 42개
# ....
# 이런 식으로 해석을 할 수 있다.
```

```python
# 다양한 set 들
a1.hist(tips['total_bill'])
a1.set_title('Tips total bill')
a1.set_xlabel('Frequency')
a1.set_ylabel('total bill')

```

### 산점도
----

* `a1.scatter(1변량데이터1, 1변량데이터2)` : 산점도 그리기

* 두 변수 간의 관계를 볼 수 있는 그래프
* 두 변수를 쓰므로 = 이변량 그래프 이다.

```python
fig1 = plt.figure()
a1 = fig1.add_subplot(1,1,1)
a1.scatter(tips['total_bill'],tips['tip'])
a1.set_title('scatterplot of total bill vs tip')
a1.set_xlabel('total bill')
a1.set_ylabel('tip')
```

### 박스 그래프
----
> 박스 그래프 설명
>* 데이터의 분포와 이상치를 함께 볼 수 있는 그래프이다.
>* 이산형 변수(뚝뚝 끊어진 값) 와 연속형 변수(연속된 숫자)를 모두 사용할 수 있는 그래프
>* 이상치 = 3사분위수 + IQR 이상 || 1사분위수 - IQR 이하
>* IQR = 3사분위수 - 1사분위수
>* n분위수 = 연속형 데이터를 값의 크기별로 정렬 후 n등분했을 때의 각 [n=1 -> n=n] 번째 순서에 있는 값

* `DF.boxplot(데이터;두 개 이상일 경우 리스트로 묶어 넣기)` : 박스 그래프 그리기

```python
import seaborn as sns
tips = sns.load_dataset('tips')

fig2 = plt.figure()
a1 = fig2.add_subplot(1,1,1)

a1.boxplot([tips[tips['sex'] == 'Male']['tip'], tips[tips['sex'] == 'Female']['tip']]
          , labels=['Female', 'Male'])
a1.set_title('tips by sex')
a1.set_xlabel('sex')
a1.set_ylabel('tips')
```

### color 이용하여 표현하기

```python
# 사용 자료 : tips
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
# 함수 선언 : femail 0 male 1로 반환
def recode_sex(sex):
    if sex == 'Femail' :
        return 0
    else :
        return 1
```

```python
# sex_num 이라는 컬럼을 하나 만들고, 그곳에 recode_sex 함수를 통한 값을 넣는다.
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips
```

```python
fig3 = plt.figure()
a1 = fig3.add_subplot(1,1,1)

a1.scatter(tips['total_bill'],
           tips['tip'],
           s=tips['size']*10,
           c=tips['sex_color'],
           alpha=0.5)
a1.set_title('total bill vs tip')
a1.set_xlabel('sex')
a1.set_ylabel('tips')

# size에 따라 크기 조절
# sex_color 값에 따라 색깔을 다르게 했고
# 투명도를 0.5로 조정
```

* seaborn 은 matplotlib 에 기반을 둔 라이브러리이다.
* matplotlib에서 나아가, 자동으로 축 제목을 잡아주는 등의 편의기능이 기본적으로 적용된다.

## 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset('tips')

sns.histplot(tips['total_bill'])


print("=== seaborn 기본 histplot ===")
ax = plt.subplots()
ax = sns.histplot(tips['total_bill'])
ax.set_title('total bill histgram')

print("=== histplot 에 선그래프 추가 ===")
ax = plt.subplots()
ax = sns.histplot(tips['total_bill'], kde=True) # 선그래프 추가
ax.set_title('total bill histgram')
```

## countplot

* `countplot(data='데이터프레임', x=구분할 범위 칼럼), order=[x축 원하는 순서])` : countplot 그래프 만들기
* 음... 히스토그램과 뭐가 다른거지?

```python
# countplot 만들기
import seaborn as sns
tips = sns.load_dataset('tips')

ax = plt.subplots()
ax = sns.countplot(data=tips, x='day')
ax.set_title('total bill countplot')
```

```python
# 히스토그램으로 동일하게 해보면?
sns.histplot(data=tips, x='day')
```

```python
# 데이터를 내가 원하는 순서대로 나오도록 하기

ax = plt.subplots()
ax = sns.countplot(data=tips, x='day', order=['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun'])
ax.set_title('total bill countplot')
```
