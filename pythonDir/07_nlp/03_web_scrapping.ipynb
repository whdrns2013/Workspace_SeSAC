## 심화 사용

````python
from bs4 import BeautifulSoup
# 간단한 페이지 만들기
test_html = '''
<html>
<head></head>

<body>
<h1 id = 'title'> 새싹반 </h1>

<div>
    <ul class='name'>
        <li><a class = 'to_naver' href='https://www.naver.com'>네이버로 갈래요</a></li>
    </ul>

    <ul class='brand'>
        <li><a href='https://www.google.com'>구글로 갈래요</a></li>
        <li><a href='https://www.youtube.com'>유튜브로 갈래요</a></li>
    </ul>
</div>
</body>
</html>
'''
````

````python
# 뷰티플슾에 넘기기
soup = BeautifulSoup(test_html, 'html.parser')
# a태그 붙은 영역의 attribution 보기
soup.a.attrs
````

### attribution 출력

````python
# 특정 영역에서 특정 attribution의 값을 가지는 항목 추출하기
# string 추출
res = soup.find('a', attrs={'class':'to_naver'})
print(res)
print(res.string)
````

````python
# 특정 영역에서 특정 attribution의 값을 가지는 항목 추출하기
# text 추출
res = soup.find('a', attrs={'class':'to_naver'})
print(res)
print(res.text)
````

### 임의의 페이지에서 데이터 가져오기

````python
# 표에서 데이터 가져오기 함수 지정

def get_text(page_num):
    from bs4 import BeautifulSoup
    import urllib.request
    
    region = []
    name = []
    address = []
    phone = []
    
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_num}&sido=&gugun=&store='
    html = urllib.request.urlopen(url)
    
    soup = BeautifulSoup(html, 'html.parser')
    tbody = soup.find_all('tbody')
    trs = tbody[0].find_all('tr')
    
    for i in trs:
        tds = i.find_all('td')
        region.append(tds[0].string)
        name.append(tds[1].string)
        address.append(tds[3].string)
        phone.append(tds[5].string)
    
    return region, name, address, phone
````

````python
# 함수 수정 : 1페이지부터 시작해, 페이지 늘려가며 스크래핑
# 또한, 표에 내용이 없으면 멈춤

def get_text(end_page):
    from bs4 import BeautifulSoup
    import urllib.request
    
    region = []
    name = []
    address = []
    phone = []
    page_num = 1
    
    for i in range(1, end_page+1):
        url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_num}&sido=&gugun=&store='
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')
        tbody = soup.find_all('tbody')
        page_num += 1

        for n in tbody:
            trs = n.find_all('tr')
            if trs[0].text.replace('\n','') == '등록된 지점이 없습니다.':
                print('스크래핑 종료')
                break
            else:
                for i in trs:
                    tds = i.find_all('td')
                    region.append(tds[0].string)
                    name.append(tds[1].string)
                    address.append(tds[3].string)
                    phone.append(tds[5].string)
    
    return region, name, address, phone
````

### (결과) 데이터프레임으로 출력

````python
# 함수 수정 : 데이터프레임으로 출력

def get_text(end_page):
    from bs4 import BeautifulSoup
    import urllib.request
    import pandas as pd
    from tqdm import tqdm
    
    region = []
    name = []
    address = []
    phone = []
    page_num = 1
    
    for i in tqdm(range(1, end_page+1)):
        url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_num}&sido=&gugun=&store='
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')
        tbody = soup.find_all('tbody')
        page_num += 1

        for n in tbody:
            trs = n.find_all('tr')
            if trs[0].text.replace('\n','') != '등록된 지점이 없습니다.':
                for i in trs:
                    tds = i.find_all('td')
                    region.append(tds[0].string)
                    name.append(tds[1].string)
                    address.append(tds[3].string)
                    phone.append(tds[5].string)
            else:
                print('스크래핑 종료')
                break
    
    df = pd.DataFrame([region, name, address, phone]).T
    df.columns = ['local', 'store_name', 'address', 'phone']
    
    return df
````

````python
# 함수 수정 : 데이터프레임으로 출력

def get_text_to_sql(end_page):
    from bs4 import BeautifulSoup
    import urllib.request
    import pandas as pd
    from tqdm import tqdm
    
    region = []
    name = []
    address = []
    phone = []
    page_num = 1
    
    ### 이부분이 sql 연결
    import pymysql as m
    con = m.connect(host = 'localhost',  db = 'hollys', 
                    user = 'root', password = '1234', charset = 'utf8')
    cur = con.cursor()
    q_id = 0
    
    for i in tqdm(range(1, end_page+1)):
        url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_num}&sido=&gugun=&store='
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')
        tbody = soup.find_all('tbody')
        page_num += 1

        for n in tbody:
            trs = n.find_all('tr')
            if trs[0].text.replace('\n','') != '등록된 지점이 없습니다.':
                for i in trs:
                    tds = i.find_all('td')
                    region.append(tds[0].string)
                    name.append(tds[1].string)
                    address.append(tds[3].string)
                    phone.append(tds[5].string)
                    
                    ### 이부분이 db에 넘기는 부분
                    sql = 'insert into hollys0111 values(%s, %s, %s, %s, %s);'
                    q_id = q_id
                    q_local = tds[1].string
                    q_store_name = tds[1].string
                    q_address = tds[3].string
                    q_phone = tds[5].string
                    cur.execute(sql, (q_id, q_local, q_store_name, q_address, q_phone))
                    con.commit()
                    q_id += 1
            else:
                print('스크래핑 종료')
                break
    
    ### db 닫기
    con.close()
    
    df = pd.DataFrame([region, name, address, phone]).T
    df.columns = ['local', 'store_name', 'address', 'phone']
    
    return df
````

## 함수화

````python
# 스크래핑 함수화

def get_text_cb(idx):
    from selenium import webdriver
    from bs4 import BeautifulSoup
    import time
    
    wd = webdriver.Chrome('chromedriver108_win.exe')
    wd.get('https://www.coffeebeankorea.com/store/store.asp')
    time.sleep(1)
    wd.execute_script(f"storePop2('{idx}')")
    time.sleep(3)
    html = wd.page_source
    
    soupCB = BeautifulSoup(html, 'html.parser')
    
    # print(soupCB.select('div.store_txt>h2'))
    # print(soupCB.select('div.store_txt>h2'))
    store_name = soupCB.select('div.store_txt>h2')[0].string
    
    store_info = soupCB.select('tbody>tr>td')
    work_time = store_info[0].text
    parking = store_info[1].text
    address = store_info[2].text
    phone = store_info[3].text
    
    ls = [idx, store_name, work_time, parking, address, phone]
    
    wd.close()
    
    return ls
````

````python
# 테스트
ls = get_text_cb(31)
ls
````

## 시맨틱 웹

http://terms.tta.or.kr/dictionary/dictionaryView.do?subject=%EC%8B%9C%EB%A7%A8%ED%8B%B1+%EC%9B%B9
