## 프로그래밍에서 기본적인 서치 메서드의 한계

파이썬에서 기본적인 서치를 진행해보자.  
우리가 알고 있는 서치는 .find 혹은 .find_all 이 기본적이다.  

```python
str1 = 'This final class with HJ'
print('정확한 서치 : ', str1.find('class'))
print('부정확한 서치 : ', str1.find('cless'))
```

하지만 위와 같이 부정확한 서치는 결과물이 나오지 않게 된다.  

SQL에서는 비교적 유연한 서치가 가능하다.  

예를 들어 SQL에서..
COL : kim, lee, park, bark ... 성씨

* 정확한 Park씨 찾기 :  
select * from ~ where col1 = 'park'  

* 박씨는 여러 표기법이 있다. Park, Bark...:  
* 이를 모두 찾기 위해 비교적 유연한 서치가 가능하다.  
select * from ~ where col1 like '*ark'  

* 하지만 이것은 ark가 들어가는 것을 찾을 뿐, 박씨를 찾는 게 아니다.  
* 거기에 'ark 가 공통으로 들어간다' 라는 점을 미리 알고 있어야 서치가 가능하다.  


여기서 참고  

* DB의 종류  
RDBMS : Mysql, Maria-DB, Oracle DB (table - based)  
객체지향 DB : NoSQL : Json, MongoDB(non-table based)  

* NoSQL이 되는 것은 무엇이 있을까?  
Json 형태가 바로 NoSQL의 대표적인 형태이다.  
Json 형태는 표가 정해져있지 않고, 키 - 밸류라는 형태만 정의되어 있다.  
Mongo DB 또한 NoSQL이며, 이는 면접 전 살펴보기를 바란다.  

* 예를 들어  
어떤 사람을 표현한다고 했을 때, 사람들의 특성은 매우매우 다르다.  
예로 어떤 사람은 전자공학, 어떤 사람은 환경공학, 어떤 사람은 음악 전공이라고 하면  
각각의 전공에서 필요한 항목들은 매우매우 다를 것이다.  
이 때문에 각각의 객체에 맞게끔 데이터베이스를 꾸리는 게 필요할 것이고,  

이것이 바로 '객체지향 데이터베이스' 즉, NoSQL 의 형태인 것이다.

맥 설정 -> 개인정보 보호 및 보안 -> 개발자 도구 -> 터미널 켜기  
  

```python
GET _search
{
  "query": {
    "match_all": {}
  }
}

###########################기본 사용법############################


get _cat/indices
# 메모리에 올라가있는 것들을 보여줌

put index1
# index1 이라는 인덱스를 생성


put index2/_doc/1
{
  "name":"mike",
  "age":25,
  "gender":"male"
}
# put : 데이터를 넣어줌
# index2라는 인덱스(그릇)를 생성하고
# document(레이블) 1번을 생성하고,
# 레이블에 name, age, gender 라는 property를 생성하고, 값을 넣어줌


get index1
# index1에 대한 정보를 불러옴

get index2
# index2에 대한 정보를 불러옴

get _cat/indices
# 메모리에 올라가있는 것들을 보여줌

put index2/_doc/2
{
  "name":"cathy",
  "old":34,
  "sex":"female",
  "hobby":"napping"
}
# Mike에 없던 항목들도 넣었고,  
# Mike에 있던 항목들 일부는 넣지 않았다.
# 엘라스틱 서치가 RDBMS와 다른 점이 바로 이것
# 칼럼이 3개, 혹은 4개로 딱 정해진 것도 아니고 
# 1번 레이블에는 칼럼이 3개 (name, age, gender)가 있고
# 2번 레이블에는 칼럼이 4개 (name, old, sex, hobby)가 있는 것이다.
# NA 값이 없다.


get index2


delete index2/_doc/2
# 지울때는 delete


get index2/_doc/2
# index2 번 인덱스의 2번 doc 읽기

put index2/_doc/2
{
  "name":"cathy",
  "old":34,
  "sex":"female",
  "hobby":"napping"
}
# 다시 cathy 넣어주고

get index2/_search
# 그러면 index2에 있는 모든 것을 검색해보자!


# in sql : insert select update delete
# in elk : put    get    (x)    delete
# 도큐먼트는 수정이 안된다 === > put으로 덮어쓰기

# 왜냐, 
# 엘라스틱서치는 "분산형 데이터베이스" 이기 때문
# 물리적으로는 여기 저기에 데이터가 흩어져있지만
# 논리적으로는 하나의 데이터베이스에 모두 있는 것으로 묶여있는데
# 이 상황에서, 어떠한 것을 '수정' 하기 위해서 그 데이터를 찾으려면
# 아주 많은 시간이 필요하게 된다. 
# 이 때문에 '수정 대신 새로 쓰고 말지'가 되기 때문에
# update를 지원하지 않고, 덮어쓰기로 대체되는 것이다.

# 그런데 이러한 '분산형' 이라는 특성 때문에
# AWS와 같은 클라우드 환경을 제공하거나 그러한 환경이 필요한 업체는 좋아하는 것이다.
# "물리적으로 데이터베이스가 나뉘어져있어도 괜찮은 방법이 있다"
# 라고도 생각할 수 있기 때문

# 그런데 '안된다'는 아니다.
# 수정보다 덮어쓰기가 자원상 효율적이기 때문에 하는 것이지
# 수정이 불가능하다는 것은 아니다.


put index2/_doc/1
{
  "name":"park",
  "age":25,
  "gender":"male"
}
# put 된 자료에 덮어쓰기를 진행하면
# version 정보가 1에서 2로 변경된다.
# 즉, 두 번째 데이터라는 것 (한 번 변경되었다는 것)


POST index2/_update/1
{
  "doc":{
    "name":"lee"
  }
}
# 업데이트도 가능하다. 
# POST와 _update 명령어를 이용하면 되며
# 이 때에도 version은 +1이 된다.
# 하지만 이러한 방식은.. 업체에 가면 혼난다.
# 혼나면서 배워야지 뭐
# POST _update 는 put 보다 훨씬 많은 시간과 리소스가 사용된다.


# get : 서버가 요청사항을 get 했다. -> 서버가 주체적으로 하는 구조
# POST : 사용자가 데이터를 수정하고자 한다. -> 서버가 수동적으로 받아들이는 구조
# get과 POST는 주어가 다르다.
# 서버가 주체가 되어 움직이는 경우가 '필요없는 것은 안하고..' 효율적으로 돌릴 수 있다.
# 주소창에 보이면 get 안보이면 POST 라고 구분하는 게 아니라, 이게 원리가 뭐인지를 알아야 한다. 
# 가끔 보면 프론트쪽에서 주소창에 정보가 보이는 것을 원치 않아서 POST 형식을 고집하는 경우가 있는데, 이러면 백엔드쪽에서는 골치가 아프다. get과 POST가 단순히 보이고 안보이고의 차이가 아니라, 자원의 효율적인 사용에 차이가 있는 것인데..


# 업데이트가 많이 필요한 서비스다?
# -> 엘라스틱서치 서버를 선택하면 안된다.
# RDBMS가 좋다.


###########################매핑############################

get index2
# get을 보면, 각 property들의 속성이 정의되어있다.
# 다이나믹 매핑 : 자동으로 type을 결정
# 명시적 매핑 :   수동으로 type을 결정
# 위는 모두 다이나믹 매핑이었다.

put index3
{
  "mappings":
  {"properties":{
    "name":{
      "type":"text"
    },
    "age":{
      "type":"short"
    },
    "gender":{
      "type":"keyword"
    }
  }
  }
}
# 각 property들의 type을 지정해주는 것

get index3
# index3을 살펴보면, 각각의 property들의 type이 정해져있음을 볼 수 있다.


###########################서치############################


put text_index/_doc/1
{
  "contents":"beautiful day"
}


get text_index/_search
{
  "query":{
    "match":{
      "contents":"beautiful day"
    }
  }
}
# 스코어가 0.5753642가 나왔다.


get text_index/_search
{
  "query":{
    "match":{
      "contents":"day"
    }
  }
}
# 스코어가 0.2876821이 나왔다.


get text_index/_search
{
  "query":{
    "match":{
      "contents":"beautif"
    }
  }
}
# 찾아지지 않는다.


get text_index/_search
{
  "query":{
    "match":{
      "contents":"Day"
    }
  }
}
# 스코어가 0.2876821이 나왔다.


get text_index/_search
{
  "query":{
    "match":{
      "contents":"wonderful day"
    }
  }
}
# 스코어가 0.2876이 나왔다.


get text_index/_search
{
  "query":{
    "match":{
      "contents":"what a beautiful day"
    }
  }
}
# 스코어가 0.2876이 나왔다.



###########################매핑 테스트############################

put test/_doc/1
{
  "contents":"what a wonderful day"
}


get test/_search
{
  "query":{
    "match":{
      "contents":"what a wonderful day"
    }
  }
}
# 스코어 1.1507283


get test/_search
{
  "query":{
    "match":{
      "contents":"wonderful day"
    }
  }
}
# 스코어 0.5753642


get test/_search
{
  "query":{
    "match":{
      "contents":"what a wonderful"
    }
  }
}
# 스코어 0.8630463


get test/_search
{
  "query":{
    "match":{
      "contents":"what a beautiful day"
    }
  }
}
# 스코어 0.8630463



###########################키워드 매핑############################

put keyword_index
{
  "mappings":
  {
    "properties":{
      "contents":{
        "type":"keyword"
      }
    }
  }
}

# type : text -> keyword


put keyword_index/_doc/1
{
  "contents":"beautiful day"
}


get keyword_index/_search
{
  "query":{
    "match":{
      "contents":"Day"
    }
  }
}
# 키워드로 묶인 contents 는 부분 검색이 안된다.


get keyword_index/_search
{
  "query":{
    "match":{
      "contents":"beautiful day"
    }
  }
}
# 꼭 일치해야 검색이 되며


get keyword_index/_search
{
  "query":{
    "match":{
      "contents":"beautiful Day"
    }
  }
}
# 대소문자까지 일치해야 한다.
# keyword 는 '범주형 데이터'에서 사용된다 => 집계에 사용됨




###########################역 인덱싱############################

# 그 전에 Index 에 대해 알아봐야 한다.
# Index는 색인으로, "검색을 빨리 하는 것"을 목표로 한다.
# 하지만 Index를 사용하면 update, delete가 느려진다는 단점이 있다. (Index가 없을 때보다 리소스를 많이 쓰게 됨)



# 역 인덱싱이란..

# 예를 들어, 아래와 같은 doc들이 만들어졌다고 해보자
# doc1 : i love dog
# doc2 : the 10 most cute dog

# 검색엔진은 받은 단어들에 대해 tokenizing을 한다.
# love, dog, most, cute

# 그러면 검색엔진은 아래와 같이 색인 작업을 미리 진행해
# 아래와 같은 인덱스 테이블을 만들게 된다.
# word : doc no.
# love : 1
# dog : 1, 2
# most : 2
# cute : 2

# 사용자가 특정 키워드를 검색했을 때, 미리 찾아둔 검색 결과를 보여준다.
# query : cute -> doc2 (cute 찾아주세요 -> doc2 를 보여주면 됨)
# query : love -> doc1
# query : dog -> doc1, doc2

# 이것이 바로 역인덱싱이다.
# 원래 인덱싱은 번호를 매기는데부터 시작한 것에 반해
# 이러한 역 인덱싱은 형태소분석을 한 후, 그에 대한 인덱싱을 한다.
# 검색엔진들은 이러한 역 인덱싱을 진행한다.




###########################분석############################


post _analyze
{
  "analyzer":"stop",
  "text":"the 10 most cute dog"
}
# stop : 일반적인 불용어 (the, 10, at ... 등)까지 날려버리는 분석기


post _analyze
{
  "analyzer":"simple",
  "text":"the 10 most cute dog"
}
# simple : 일반적인 불용어를 날려버리지 않는 분석기







```
